```python
import pandas as pd

sample_1 = pd.read_excel('./files/sample_1.xlsx',
                        header =1,
                        skipfooter = 2,
                        usecols = 'A:C')


sample_1['기준연월'] = '2019-11'
sample_1
```

```python
code_master = pd.read_excel('./files/sample_codemaster.xlsx')
code_master
```

```python
sample_1_code = pd.merge(left = sample_1, 
                        right = code_master,
                        how = 'left',
                        left_on = '국적코드',
                        right_on = '국적코드')

sample_1_code
```

```python
sample_2 = pd.read_excel('./files/sample_2.xlsx',
                        header = 1,
                        skipfooter = 2,
                        usecols = 'A:C')
sample_2['기준연월'] = '2009-12'
sample_2_code = pd.merge(left = sample_2,
                        right = code_master,
                        how = 'left',
                        left_on = '국적코드',
                        right_on = '국적코드')

sample_2_code
```

```python
sample = sample_1_code.append(sample_2_code, ignore_index = True)
sample
```

```python
sample.to_excel('./sample.xlsx', index = False)
```
